```python
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings(action='ignore', category=ConvergenceWarning)
import time

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from tqdm import tqdm

from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error as mse
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler

%matplotlib inline
```

## Build a simple linear model regularizing with lasso with a ramdom lambda value
#### First example

```python
#Get data
data = pd.read_csv('Master_with_technicals_news.csv')

#Get ticker data
tickers = pd.read_csv("tickers_list_news.csv")['0'].values
#Create ticker dataframe
index = []
for i in range(len(tickers)):
    index.append(i)
tickers_df = pd.DataFrame({'Index':index, 'Ticker':tickers}).reset_index(drop=True)
tickers_dict = tickers_df['Ticker'].to_dict()

tickers_dict = {value: key for key, value in tickers_dict.items()}
```

```python
#split data into X and Y
X = data.drop(["date","returns_pred"], axis = 1)

X['index_ticker']=X['Ticker'].map(tickers_dict)

y = data["prev_returns"].copy()
```

```python
pipe_lasso = make_pipeline(
    PolynomialFeatures(degree=3, include_bias=False),
    StandardScaler(),
    Lasso(random_state=1),
)
```

```python
def RMSE(ytrue, ypred):
    rmse = np.sqrt(mse(ytrue,ypred))
    return rmse
```

```python
X = X.drop("Ticker", axis = 1)
X_dev, X_test, y_dev, y_test = train_test_split(X, y, test_size=1/3, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_dev, y_dev, test_size=1/2, random_state=1)
```

```python
start_time = time.time()

pipe_lasso.fit(X_train, y_train) #Train with lasso

ypred_val = pipe_lasso.predict(X_val)#predicted values for validation
ypred_test = pipe_lasso.predict(X_test)#predicted values for test

print("RMSE val:  ", RMSE(y_val,ypred_val))
print("RMSE test: ", RMSE(y_test, ypred_test))

end_time = time.time()

elapsed_time = end_time - start_time
time_min = elapsed_time/60
time_h = time_min/60

print(time_min)
```

### Validation curve

```python
X_dev, X_test, y_dev, y_test = train_test_split(X, y, test_size=1/3, random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_dev, y_dev, test_size=1/2, random_state=1)
```

```python
start_time = time.time()

from sklearn.model_selection import validation_curve
lambdas =  np.logspace(-4, -2, 40)

train_scores, test_scores = \
    validation_curve(estimator=pipe_lasso,
                     X=X_dev,
                     y=y_dev,
                     param_name='lasso__alpha',
                     param_range=lambdas,
                     scoring='neg_mean_squared_error',# scoring='neg_mean_squared_error',                 
                     cv=5)

mean_values = pd.concat({'train': pd.DataFrame(-train_scores).mean(1), 
                         'test': pd.DataFrame(-test_scores).mean(1), 
                         'lambda': pd.DataFrame(lambdas).mean(1)}, axis =1)

# plot curves
pd.concat({'train': pd.DataFrame(-train_scores).mean(1), 
           'test': pd.DataFrame(-test_scores).mean(1)},
           axis=1)\
    .pipe(np.sqrt)\
    .set_index(pd.Index(lambdas, name='lambda'))\
    .plot(logx=True, logy=True)

plt.show()

end_time = time.time()

elapsed_time = end_time - start_time
time_min = elapsed_time/60
time_h = time_min/60

print("Time elapsed(min): ",time_min)
```

### Cross validation

```python
from sklearn.model_selection import KFold
import tqdm

start_time = time.time()

X_dev, X_test, y_dev, y_test = train_test_split(X, y, test_size=1/3, random_state=1)
# split development into test and validation
#X_train, X_val, y_train, y_val = train_test_split(X_dev, y_dev, test_size=1/2, random_state=1)

kfolds = KFold(n_splits=5)
folds = list(kfolds.split(X_dev, y_dev))
lambdas = np.logspace(-4,4,12)


MSE = []
for i in tqdm.tqdm(range(len(lambdas))):
    MSE_ = []
    for train_idx, val_idx in folds:
        Lasso_pipe = make_pipeline(PolynomialFeatures(degree = 3, include_bias = False),
                      StandardScaler(),
                      Lasso(alpha = lambdas[i]))
        #Create dataset into folds
        X_train, y_train = X_dev.iloc[train_idx], y_dev[train_idx]
        X_val, y_val = X_dev.iloc[val_idx], y_dev[val_idx]
        #fit the folded training sets
        Lasso_pipe.fit(X_train, y_train)
        
        #predict with validationsets and find mse
        ypred_val = Lasso_pipe.predict(X_val)
        MSE_.append(mse(y_val, ypred_val))
    MSE.append(np.mean(MSE_))

end_time = time.time()

elapsed_time = end_time - start_time
time_min = elapsed_time/60
time_h = time_min/60

print("Time elapsed(min): ",time_min)
```
